### 3. Visualize the Graded Result

You can define input and output video paths in the following cell, then run it to visualize the result.

```python
from IPython.display import HTML
from base64 import b64encode

input_path = 'input/input-preview.mp4'
output_path = 'output/output-preview.mp4'

def show_video(input_path, output_path, video_width = 1200):
    input_video_file = open(input_path, "r+b").read()
    input_video_url = f"data:video/mp4;base64,{b64encode(input_video_file).decode()}"
    output_video_file = open(output_path, "r+b").read()
    output_video_url = f"data:video/mp4;base64,{b64encode(output_video_file).decode()}"
    return HTML(f"""
        <div style="display: flex;">
            <div style="flex: 1;">
                <video width={video_width} controls><source src="{input_video_url}"></video>
            </div>
            <div style="flex: 1; margin-left: 10px;">
                <video width={video_width} controls><source src="{output_video_url}"></video>
            </div>
        </div>
    """)

show_video(input_path, output_path)
```
